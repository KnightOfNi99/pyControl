Create random Data in Python

```python
age = np.random.uniform(0,18,size=50)
height = 22 + 8.6 * age + np.random.normal(size=50,scale=10)

growing=pd.DataFrame({'age':age,
                     'height':height})
growing.head()
```

## Application : Simple regression

Create linear regression model using `age` and `height` var defined in python

```python
%%R -i age,height -o coefficients

fit <- lm(height~age)
coefficients <- coef(fit)
```

```python
print(coefficients)
```

However, `coefficients` is till an R object.

```python
coefficients
```

Convert to python

```python
pandas2ri.ri2py(coefficients)
```

```python
%%R

summary(fit)  # output in cmd window
```

Generate some diagnostic plots for the linear regression model

```python
%%R -w 900 -h 450

par(mfrow=c(1,2),cex=1.25)

plot(fit,which=1:2)
```

Generate scatter plot superimposed linear regression fit (and confidence intervals)

```python
%%R -w 900 -h 400 -i growing

library(ggplot2)

ggplot(growing, aes(x=age, y=height)) +
    geom_point(size=3, color='blue') +
    geom_smooth(method='lm', color='red',lty='dashed') +
    theme_classic(base_size=16)
```
